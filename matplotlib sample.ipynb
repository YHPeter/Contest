## Import library

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
```

## Define iterating function of dots

```python
def movingDots(n,x,y,z,dot):
    """
    Generate the new x,y,z coordinates of dots
    Change the visual angle
    """
    print(x[n],y[n],z[n])
    dot.set_data(x[n],y[n])
    dot.set_3d_properties(z[n],'z',)
    # ax.elev += 1
    ax.azim += 0.0072*200
    return dot
```

## Set figure style

```python
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(111, projection='3d')      # set 3D image
ax.set_title('Constrained Maximization')        # set title
ax.set_xlabel('Plant A')                        # set xlabel
ax.set_ylabel('Plant B')                        # set ylabel
ax.set_zlabel('Total Cost /$10^{7}$')           # set zlabel
dot, = ax.plot([],[],[], 'r.')                  # set style of dot 'red dot'
```

## Plot 3D surface

```python
X, Y = np.meshgrid(np.arange(50,310,1), np.arange(50,310,1))    # generate grid 
Z = np.array(120*X**2+60*Y**2-20*X*Y)/10**7                     # according to the formula, generate Z values
ax.plot_surface(X, Y, Z, cmap = plt.get_cmap('Accent'))         # plot surface, style: 'Accent'
```

## Plot line

```python
x,y = np.arange(50,310,1), np.arange(50,310,1)  # set range of line
z = np.array(120*x**2+60*y**2-20*x*y)/10**7     # generate z values of the line
ax.plot3D(x,y,z,'black')                        # plot line, style: 'black'
```

```python
ani = FuncAnimation(fig, movingDots, fargs=(x,y,z,dot),interval=50, frames=250) 
```
